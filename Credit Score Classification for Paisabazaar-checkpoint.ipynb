```python
# 5. Hypothesis Testing
## Hypothetical Statement - 1
# H0: Income does not affect credit score
# H1: Income does affect credit score
from scipy.stats import f_oneway
credit_categories = df['Credit_Score'].unique()
groups = [df[df['Credit_Score'] == cat]['Annual_Income'] for cat in credit_categories]
print(f_oneway(*groups))
```

```python
## Hypothetical Statement - 2
# H0: Credit card usage does not vary by credit score
# H1: Credit card usage varies by credit score
groups = [df[df['Credit_Score'] == cat]['Credit_Utilization_Ratio'] for cat in credit_categories]
print(f_oneway(*groups))
```

```python
## Hypothetical Statement - 3
# H0: There is no relationship between late payments and credit score
# H1: There is a relationship
sns.boxplot(x='Credit_Score', y='Num_of_Delayed_Payment', data=df)
```

```python
# 6. Feature Engineering & Data Preprocessing
## Label Encoding for categorical columns
le = LabelEncoder()
df['Occupation'] = le.fit_transform(df['Occupation'])
```

```python
## Feature Selection
# Select only numeric data for correlation
numeric_df = df.select_dtypes(include=np.number)
corr = numeric_df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python
## Data Scaling
scaler = StandardScaler()
# Ensure only numeric features are used for scaling
features = df.drop('Credit_Score', axis=1).select_dtypes(include=np.number)
# Encode target variable for modeling
target = LabelEncoder().fit_transform(df['Credit_Score'])
X_scaled = scaler.fit_transform(features)
```

```python
## Data Splitting
X_train, X_test, y_train, y_test = train_test_split(X_scaled, target, test_size=0.3, random_state=42, stratify=target)
```

```python
## Handling Imbalanced Dataset
smote = SMOTE()
X_train, y_train = smote.fit_resample(X_train, y_train)
```

```python
# 7. ML Model Implementation
## Model 1: Logistic Regression
model1 = LogisticRegression()
model1.fit(X_train, y_train)
y_pred1 = model1.predict(X_test)
print(classification_report(y_test, y_pred1))
```

```python
## Model 2: Random Forest
model2 = RandomForestClassifier()
model2.fit(X_train, y_train)
y_pred2 = model2.predict(X_test)
print(classification_report(y_test, y_pred2))
```

```python
## Model 3: XGBoost
model3 = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
model3.fit(X_train, y_train)
y_pred3 = model3.predict(X_test)
print(classification_report(y_test, y_pred3))
```

```python
## Hyperparameter tuning for best model
params = {'n_estimators': [100, 200], 'max_depth': [3, 5, 7]}
gs = GridSearchCV(RandomForestClassifier(), param_grid=params, cv=5)
gs.fit(X_train, y_train)
y_pred_gs = gs.predict(X_test)
print(classification_report(y_test, y_pred_gs))
```
